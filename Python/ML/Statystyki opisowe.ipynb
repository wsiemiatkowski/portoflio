### 3. Podstawowe statystyki opisowe
#### 🇬🇧 Descriptive statistics

Możemy w łatwy sposób przedstawić podstawowe statystyki dla naszych danych.

Dużo informacji jesteśmy w stanie uzyskać przy wykorzystaniu analizy korelacji. Możemy dzięki niej uzyskać wgląd w ciekawe zależności międy danymi i odkryć prawidłowości interesujące z naukowego lub biznesowego punktu widzenia.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
# importujemy biblioteki, które przydadzą nam się w analizie korelacji
```

```python
df.corr()
# przy użyciu tego prostego kodu uzyskujemy współczynniki korelacji między naszymi zmiennymi
```

```python
sns.heatmap(df.corr())
# jeżeli zależy nam na uzyskaniu ciekawszych wizualizacji możemy skorzystać z biblioteki seaborn
```

```python
plt.figure(figsize=(10, 5))
# określamy wielkość tabeli, którą generujemy
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True)
# odwołujemy się do metody tj. mapy ciepła
heatmap.set_title('Analiza korelacji', fontdict={'fontsize':14}, pad=12)
# ustalamy tytuł tabeli
```

Sam współczynnik korelacji nie jest jeszcze wszystkim. Potrzebujemy również informacji dotyczącej istotności statystycznej (p). Standardowo te współczynniki korelacji, które są mniejsze od 0,05 uznaje się za wartości istotne statystycznie.

```python
r = df.corr()
# na początek zapisujemy nową zmienną, która jest tabelą korelacji
p = df.corr(method=lambda x, y: pearsonr(x, y)[1]) - np.eye(*r.shape)
# następnie, nie wchodząc w szczegóły, definiujemy wartość p
p = p.applymap(lambda x: ''.join(['*' for t in [0.01,0.05,0.1] if x<=t]))
r.round(2).astype(str) + p
# nakładamy wartość p na tabelę korelacji
```
