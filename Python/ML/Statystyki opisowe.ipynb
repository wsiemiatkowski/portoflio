```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
# importujemy biblioteki, kt贸re przydadznam siw analizie korelacji
```

```python
df.corr()
# przy u偶yciu tego prostego kodu uzyskujemy wsp贸czynniki korelacji midzy naszymi zmiennymi
```

```python
plt.figure(figsize=(10, 5))
# okrelamy wielkotabeli, kt贸r generujemy
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True)
# odwoujemy sido metody tj. mapy ciepa
heatmap.set_title('Analiza korelacji', fontdict={'fontsize':14}, pad=12)
# ustalamy tytu tabeli
```

Sam wsp贸czynnik korelacji nie jest jeszcze wszystkim. Potrzebujemy r贸wnie偶 informacji dotyczcej istotnoci statystycznej (p). Standardowo te wsp贸czynniki korelacji, kt贸re smniejsze od 0,05 uznaje siza wartoci istotne statystycznie.

```python
r = df.corr()
# na pocztek zapisujemy nowzmienn, kt贸ra jest tabel korelacji
p = df.corr(method=lambda x, y: pearsonr(x, y)[1]) - np.eye(*r.shape)
# nastpnie, nie wchodzc w szczeg贸y, definiujemy warto p
p = p.applymap(lambda x: ''.join(['*' for t in [0.01,0.05,0.1] if x<=t]))
r.round(2).astype(str) + p
# nakadamy wartop na tabelkorelacji
```
