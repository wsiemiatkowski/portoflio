# **Dynamika ruchu po równi pochyłej**

### Ruch postępowy środka masy:

> ### (1) $\quad m a = mg \sin(\alpha) - T $,

gdzie:

 **m** to masa staczającej się bryły,
 
 **$\alpha$** - kąt nachylenia równi,
 
 **g** - przyspieszenie ziemskie,
 
 **T** - siła tarcia posuwistego przyłożona w punkcie styczności kulki z podłożem równi i skierowana wzdłuż równi ku górze.

### Ruch obrotowy kuli lub walca:

> ### (2) $\quad I \varepsilon = TR$,

gdzie: 

**I** - moment bezwładności bryły względem osi przechodzącej przez środek masy,

R - promień kuli/walca,

$\varepsilon = \dot{\omega} = \ddot{\phi}$ - przyspieszenie kątowe .

#### Z powyższego układu równań można wyznaczyć wzory na przyspieszenie liniowe $a$ i kątowe $\varepsilon$. Wielkości te zależą jedynie od parametrów równi i bryły, a więc nie zależą od czasu:

> ## (i)  $\quad a = g \frac{ \sin\alpha}{1 + \frac{I}{mR^2}}$

> ## (ii) $\quad \varepsilon = a/R$

# **Moment bezwładności**

Moment bezwładności wyraża opór stawiany przez ciało przy zmianie prędkości obrotowej, dlatego jego wartość zależy od wyboru osi obrotu.

 # **Przykład - moment bezwładności obręczy** 

 Wyprowadź wzór na moment bezwładnośći względem osi przechodzącej przez środek ciężkości **obręczy** o promieniu R i masie m. 

 image.png

$ I = \sum m_i (x_i^2 + y_i^2) = \sum m_i R^2 = R^2 \sum m_i =  R^2 \,m$

 # **Przykład - moment bezwładności płaskiego krążka**
 
Wyprowadź wzór na moment bezwładnośći względem osi przechodzącej przez środek ciężkości **płaskiego krążka** o promieniu R i masie m. 
 
 image.png

Dzielimy krążek na cienkie pierścienie o promieniu r i szerokości dr.

Element masy pierścienia:

> $ dm = m  \frac{\text{pole wycinka}}{\text{pole krążka}} = m \frac{2 \pi r \, dr}{\pi R^2} = \frac{2mr}{R^2}dr$

Moment bezwładności pierścienia:

> $I_p = r^2 dm = \frac{2m}{R^2} r^3 dr$

Moment bezwładnośći krążka = $\sum$ moment bezwładności pierścieni:

> $I = \sum I_p, \quad $

 czyli dla ciągłego rozkładu masy:

> $I = \int_0^R \frac{2m}{R^2}  r^3 dr = \frac{2m}{R^2}\frac{R^4}{4} = \frac{1}{2}mR^2$



 # **Przykład** - Prędkość na dole równi wyliczona z zasady zachowania energii

 Walec ($I_w = mR^2/2$) i kula ($I_k = \frac{2}{5}mR^2$) staczają się po równi pochyłej o wysokości $h$. Policz wartość prędkości liniowej na dole równi w przypadku, gdy tarcie wykonuje pracę (ruch obrotowo-postępowy) oraz w przypadku, gdy tarcie nie wykonuje pracy (ruch postępowy).

 # **Zadanie 1 (do dokończenia w domu)**

Rozwiąż równania ruchu kuli toczącej się po porówni pochyłej korzystając z metody Eulara. Przyjmij $m = 1kg$, $\alpha = 30 ^\circ $, $R = 0.5 m$, $g = 9.81 \, m/s^2$, długość równi $s = 10 \, m$.

> (a) Wyznacz prędkość liniową i kątową kuli na dole równi.

> (b) Policz prędkość liniową no dole równi przy pomocy zasady zachowania energii.

```python
import matplotlib.pyplot as plt
import numpy as np

def odeEuler(f1, f2, f3, f4, v0, x0, w0, q0, t, R, a):
  
    v = np.zeros(len(t))
    x = np.zeros(len(t))
    w = np.zeros(len(t))
    q = np.zeros(len(t))

    v[0] = v0
    x[0] = x0
    w[0] = w0
    q[0] = q0

    for n in range(0,len(t)-1):
        h = t[n+1] - t[n]
        v[n+1] = v[n] + f1(v[n], a) * h
        x[n+1] = x[n] + f2(x[n], v[n+1]) * h
        w[n+1] = w[n] + f3(w[n], a, R) * h
        q[n+1] = q[n] + f4(q[n], w[n+1]) * h

        if x[n+1] > 10: #nie ma dokładnego x == 10 więc daje przedział
          print(f"V na końcu równi {v[n+1]}")
          print(f"W na końcu równi {w[n+1]}")
         
    return v, x, w, q

a = 0
b = 3
n = 100
t = np.linspace(a, b, n)

v0 = 0
x0 = 0
w0 = 0
q0 = 0

m = 1
alpha = np.pi / 6
R = 0.5
g = 9.81
s = 10

def f1(v, a):
  return a

def f2(x, v):
  return v

def f3(w, a, R):
  return a / R

def f4(q, w):
  return w

def f5(g, alpha, m, s, R):
  return g * np.sin(alpha) / (1 + (2/5)) #I/mR^2 = 2/5 * mR^2/mR^2 = 2/5

a0 = f5(g, alpha, m, s, R)

v, x, w, q = odeEuler(f1, f2, f3, f4, v0, x0, w0, q0, t, R, a0)


plt.figure(1)
plt.title("Kula")

plt.plot(t, v,'bo',label = 'v')
plt.plot(t, x,'go',label = 'w')
plt.plot(t, w,'mo',label = 'x')
plt.plot(t, q,'ro',label = 'q')

plt.xlabel('t', fontsize =14)
plt.ylabel('Zmienne', fontsize =14)

plt.legend()
plt.show()

vE = np.sqrt(10/7 * g * 1/2 * s) # sin(alpha) = h/s => h = s * sin(30) = 1/2 * s

print(f"V na końcu równi z zasady zakochania energi = {vE}")
```
